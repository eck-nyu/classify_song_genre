### Import libraries, song track data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.model_selection import KFold, cross_val_score
from sklearn.pipeline import Pipeline

# Read in track metadata with genre labels
tracks = pd.read_csv("fma-rock-vs-hiphop.csv");

# Read in track metrics with the features
echonest_metrics = pd.read_json("echonest-metrics.json", precise_float=True);

# Merge the relevant columns of tracks and echonest_metrics
echo_tracks = echonest_metrics.merge(tracks[['track_id', 'genre_top']], on='track_id');

# Inspect the resultant dataframe
echo_tracks.info()
```

### Initial check: Plot correlation matrix to see if there are obvious redundancies between variables for freebie reduced dimensions 

```python
# Create a correlation matrix
corr_metrics = echo_tracks.corr()
corr_metrics.style.background_gradient()
```

## Initialize one splitting of the dataset into training, test data 

```python
# Create features
features = echo_tracks.drop(["genre_top","track_id"], axis=1).values

# Create labels
labels = echo_tracks["genre_top"]

# Split our data
train_features, test_features, train_labels, test_labels = train_test_split(
        features, labels, random_state=10)
print("Divided the data into training/test.")
```

```python
# Pair plots can't hurt.
# Comment out for checking/running project to save run time.
%matplotlib inline

# Create dataframe version just for input into scatter matrix 
features_plt = echo_tracks.drop(["genre_top","track_id"],axis=1) 

scatter_matrix = pd.plotting.scatter_matrix(features_plt, alpha=0.1, marker='.');
for ax in scatter_matrix.ravel():
    # Rotate the title names and turn off tick labels
    ax.set_xlabel(ax.get_xlabel(), fontsize=10, rotation=90)
    ax.set_ylabel(ax.get_ylabel(), fontsize=10, rotation=0)
    ax.xaxis.set_ticks([])
    ax.yaxis.set_ticks([])
    
plt.suptitle("Echo_tracks");
```

## Reduce dimensions with PCA. 
### First, need to rescale so mean=0, range [-1,1] 

```python
# Scale the features and set the values to a new variable, necessary for PCA 
scaler = StandardScaler()

# Scale train_features and test_features
scaled_train_features = scaler.fit_transform( train_features )
scaled_test_features = scaler.transform( test_features )
```

```python
# Reduce dimensions with PCA
pca = PCA()
pca.fit(scaled_train_features)

# Get the explained variance vector
exp_variance = pca.explained_variance_ratio_ 
```

### Plot the PCA results

```python
# Calculate the cumulative explained variance
cum_exp_variance = np.cumsum(exp_variance)
threshold = 0.90;

# plot the explained variance using a barplot
fig, ax = plt.subplots();
ax.bar(range(len(exp_variance)), exp_variance);
ax.set_xlabel('Principal Component #');
ax.set_ylabel('Explained variance');
plt.suptitle('PCA of training data');

# Plot the cumulative explained variance and draw a dashed line at 0.85.
ax2 = ax.twinx();#plt.subplots();
ax2.plot(range(len(cum_exp_variance)), cum_exp_variance, color='red');
ax2.set_ylabel("Cumulative explained variance", color='red')
ax2.axhline(y=threshold, linestyle='--', color='red');
```

### Now apply the PCA with reduced dimensions to the training and test data

```python
# Perform PCA with the chosen number of components and project data onto components

idx = np.argmax(cum_exp_variance > threshold) + 1 # Index to number of elements
print("Keeping ",str(idx)," components for threshold ",str(threshold), "variance.")
pca = PCA(n_components=idx, random_state=10)

# Fit and transform the scaled training features using pca
train_pca = pca.fit_transform(scaled_train_features)

# Fit and transform the scaled test features using pca
test_pca = pca.transform(scaled_test_features)
```

## Train a decision tree and log reg model with the dimension-reduced PCA data 

```python
# Train our decision tree
tree = DecisionTreeClassifier(random_state=10)
tree.fit(train_pca, train_labels)

# Predict the labels for the test data
pred_labels_tree = tree.predict(test_pca)

# Print out sampling list to see how tree model did
print("Some sample results to take a look at how the tree model did: ")
print("Test vals: ",test_labels.values[1:27:3])
print("Pred vals: ", pred_labels_tree[1:27:3])


# Train our logistic regression and predict labels for the test set
logreg = LogisticRegression(random_state=10, solver='lbfgs')
logreg.fit(train_pca,train_labels)

pred_labels_logit = logreg.predict(test_pca)
```

### Display the classification reports for our two models

```python
# Create the classification report for both models
target_names = ["Hip-Hop","Rock"]

class_rep_tree = classification_report(test_labels, pred_labels_tree, target_names=target_names)
class_rep_log = classification_report(test_labels, pred_labels_logit, target_names =target_names)

print("Decision Tree: \n", class_rep_tree)
print("Logistic Regression: \n", class_rep_log)
```

### See discrepancy between precision of hip-hop/rock predictions for both model... 
## Re-balance the training data: make ~equal sample sizes of each genre

```python
# Subset only the hip-hop tracks, and then only the rock tracks
hop_only = echo_tracks.loc[echo_tracks["genre_top"]=="Hip-Hop"]
rock_only = echo_tracks.loc[echo_tracks["genre_top"]=="Rock"]

hop_num = len(hop_only)
rock_num = len(rock_only)
print("Number of hip-hop tracks: ",hop_num)
print("Number of rock tracks: ",rock_num)

# sample the rocks songs to be the same number as there are hip-hop songs
rock_only = rock_only.sample( frac=hop_num/rock_num, random_state=10 )
print("Number of rock tracks after: ",len(rock_only))

# concatenate the dataframes rock_only and hop_only
rock_hop_bal = pd.concat([rock_only,hop_only])

# The features, labels, and pca projection are created for the balanced dataframe
features = rock_hop_bal.drop(['genre_top', 'track_id'], axis=1)
labels = rock_hop_bal['genre_top']

# Redefine the train and test set with the pca_projection from the balanced data
train_features, test_features, train_labels, test_labels = train_test_split(
    features, labels, random_state=10)

train_pca = pca.fit_transform(scaler.fit_transform(train_features))
test_pca = pca.transform(scaler.transform(test_features))
```

## Re-train the models using the balanced data

```python
# Train our decision tree on the balanced data
tree = DecisionTreeClassifier(random_state=10)
tree.fit(train_pca, train_labels)

# Predict the labels for the test data
pred_labels_tree = tree.predict(test_pca)

# Print out sampling list to see how tree model did
print("Some sample values to see results")
print("Test vals: ",test_labels.values[1:30:3])
print("Pred vals: ", pred_labels_tree[1:30:3])#.size

# Train our logistic regression on the balanced data
logreg = LogisticRegression(random_state=10, solver='lbfgs')
logreg.fit(train_pca,train_labels)

pred_labels_logit = logreg.predict(test_pca)


# Create the classification report for both models
target_names = ["Hip-Hop","Rock"]

class_rep_tree = classification_report(test_labels, pred_labels_tree, target_names=target_names)
class_rep_log = classification_report(test_labels, pred_labels_logit, target_names =target_names)
print(" ")
print("Class. reports for balanced models:")
print("Decision Tree: \n", class_rep_tree)
print("Logistic Regression: \n", class_rep_log)
```

## Use K-fold cross validation to use different subsets as training/testing data

```python
# First create a pipeline of the whole process once data is divided

tree_pipe = Pipeline([("scaler", StandardScaler()), ("pca", PCA(n_components=6)), 
                      ("tree", DecisionTreeClassifier(random_state=10))])
logreg_pipe = Pipeline([("scaler", StandardScaler()), ("pca", PCA(n_components=6)), 
                        ("logreg", LogisticRegression(random_state=10, solver='lbfgs'))])

# Set up our K-fold cross-validation
kf = KFold(n_splits=10)

# Train our models using KFold cv
tree_score = cross_val_score(tree_pipe, features, y=labels, cv=kf)
logit_score = cross_val_score(logreg_pipe, features, y=labels, cv=kf)

# Print the mean of each array of scores
print("Decision Tree values: ", np.round(tree_score,2))
print("Log. Reg. values: ",np.round(logit_score,2))
print("")
print("Averages: ")
print("Decision Tree Accuracy: ", np.round(100*np.mean(tree_score),1),"+/-",np.round(100*np.std(tree_score),1),"%")
print("Logistic Regression Accuracy: ", np.round(100*np.mean(logit_score),1),"+/-",np.round(100*np.std(logit_score),1),"%")
```
